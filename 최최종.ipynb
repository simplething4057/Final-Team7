```python
df['Date']= pd.to_datetime(df['Date'])
df['Production Date'] = pd.to_datetime(df['Production Date'])
df['Expiration Date'] = pd.to_datetime(df['Expiration Date'])

# 유통기한 데이터
avg_Shelf_life = df.groupby('Product Name')['Shelf Life (days)'].mean().reset_index()
avg_Shelf_life
```

총 판매량 = (기한 내 판매량 + 기한 후 판매량)

기한 내 소진율(%) = (기한 내 판매량/총 판매량)*100

본 데이터셋의 칼럼에는 판매일자는 없으므로, date 칼럼이 Expiration Date 칼럼보다 큰 일자를 기한 후 판매일자로 가정하여 판매일자를 계산


```python
# [Date < Expiration Date 조건]
Within_shelf = df[df['Date'] < df['Expiration Date']]

# [Date > Expiration Date 조건]
After_shelf = df[df['Date'] > df['Expiration Date']]
```

```python
# 제품 별 판매량 합치기
within_sales = Within_shelf.groupby('Product Name')['Quantity Sold (liters/kg)'].sum().reset_index()
within_sales.rename(columns={'Quantity Sold (liters/kg)': 'Within Shelf Life Sales'}, inplace=True)

after_sales = After_shelf.groupby('Product Name')['Quantity Sold (liters/kg)'].sum().reset_index()
after_sales.rename(columns={'Quantity Sold (liters/kg)': 'After Shelf Life Sales'}, inplace=True)
```

```python
sales_comparison = pd.merge(within_sales, after_sales, on='Product Name', how='outer').fillna(0)
```

```python
# 비율
sales_comparison['Total Sales'] = (sales_comparison['Within Shelf Life Sales'] +
                                   sales_comparison['After Shelf Life Sales'])
sales_comparison['Within Ratio (%)'] = (sales_comparison['Within Shelf Life Sales'] /
                                        sales_comparison['Total Sales'] * 100).round(2)
sales_comparison['After Ratio (%)'] = (sales_comparison['After Shelf Life Sales'] /
                                       sales_comparison['Total Sales'] * 100).round(2)
```

```python
# 재품 별 평균 재고
stock = df.groupby('Product Name')['Quantity in Stock (liters/kg)'].mean().reset_index()
stock.rename(columns={'Quantity in Stock (liters/kg)': 'Avg Stock'}, inplace=True)

sales_stock = pd.merge(sales_comparison, stock, on="Product Name", how="left")
```

```python
print("제품별 기한 내/후 판매 비율 + 평균 실제 재고")
print(sales_stock)
```
